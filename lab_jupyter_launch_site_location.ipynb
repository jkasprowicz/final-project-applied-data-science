```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

# Haversine function to calculate distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c

# Launch site coordinates (e.g., CCAFS SLC-40)
launch_site_coords = (28.563197, -80.576820)

# Proximity points (e.g., cities, ports)
proximity_points = {
    "Orlando": (28.538336, -81.379234),
    "Port Canaveral": (28.413300, -80.633900),
    "Kennedy Space Center Visitor Complex": (28.572064, -80.648015),
}

# Calculate distances
distances = {}
for location, coords in proximity_points.items():
    distances[location] = haversine(launch_site_coords[0], launch_site_coords[1], coords[0], coords[1])

# Display results
for location, distance in distances.items():
    print(f"Distance from {launch_site_coords} to {location}: {distance:.2f} km")
```

Next, we need to explore and analyze the proximities of launch sites.


```python
for location, coords in proximity_points.items():
    folium.Marker(coords, popup=f"{location}: {distances[location]:.2f} km").add_to(site_map)
    folium.PolyLine([launch_site_coords, coords], color="blue", weight=2.5).add_to(site_map)

# Display map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon
import math

# Define the Haversine function to calculate the distance
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c

# Coordinates for the launch site and closest city (for example)
launch_site_lat, launch_site_lon = 28.563197, -80.576820
city_lat, city_lon = 28.538336, -81.379234  # Coordinates for Orlando, FL (as an example)

# Calculate the distance from the launch site to the closest city
distance_to_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)

# Initialize the map centered at the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Add a marker for the launch site
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color="blue", icon="info-sign"),
).add_to(site_map)

# Add a marker for the closest city (or other feature like railway, highway, etc.)
folium.Marker(
    [city_lat, city_lon],
    popup=f"Closest City (Distance: {distance_to_city:.2f} km)",
    icon=folium.Icon(color="green"),
).add_to(site_map)

# Draw a line between the launch site and the closest city
folium.PolyLine(
    [[launch_site_lat, launch_site_lon], [city_lat, city_lon]], color="blue", weight=2, opacity=0.7
).add_to(site_map)

# Add a distance marker showing the distance between the two locations
distance_marker = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_to_city:.2f} KM</b></div>',
    ),
)
distance_marker.add_to(site_map)

# Display the map
site_map
```
